```python
import cv2
from PIL import Image
import numpy as np
import mediapipe as mp
from scipy.spatial import minkowski_distance

mpHands = mp.solutions.hands
hands = mpHands.Hands(max_num_hands=1, min_detection_confidence=0.7)
mpDraw = mp.solutions.drawing_utils
```

```python
def landmarkGen(frame, ch):
    frame = cv2.imread(frame)
    x, y, c = frame.shape
    framergb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(framergb)

    landmarks = []
    if result.multi_hand_landmarks:
        for handslms in result.multi_hand_landmarks:
            for lm in handslms.landmark:
                lmx = int(lm.x * x)
                lmy = int(lm.y * y)

                landmarks.append([lm.x, lm.y, lm.z])
    cv2.imwrite('output.jpg', frame)
    return (ch, landmarks)


def landmarkGen2(frame):
    x, y, c = frame.shape
    framergb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(framergb)

    landmarks = []
    if result.multi_hand_landmarks:
        for handslms in result.multi_hand_landmarks:
            for lm in handslms.landmark:
                lmx = int(lm.x * x)
                lmy = int(lm.y * y)

                landmarks.append([lm.x, lm.y, lm.z])

    return (landmarks)


la = landmarkGen('a.jpg', 'a')
lb = landmarkGen('b.jpg', 'b')
lc = landmarkGen('c.jpg', 'c')


def calcLandMDist(laaaa):
    allLandmark = [la, lc]
    res = []
    if len(laaaa) == 0:
        return
    for mark in allLandmark:
        for m in range(21):
            dist = minkowski_distance(mark[1][m], laaaa[m], 2)
        res.append((mark[0], dist))
    min_value = min(res, key=lambda x: x[1])
    print(min_value, res)
    return min_value[0]
```

```python
import random
def generateRandomSeq():
    seq = []
    for i in range(5):
        seq.append(random.choice(['a', 'c']))
    return seq

# Initialize the webcam for Hand Gesture Recognition Python project
cap = cv2.VideoCapture(0)
def main():
    seqn = generateRandomSeq()
    curr_seq = 0
    beat = 30
    curr = 0
    auth = False
    result = []
    while True and auth == False and curr_seq < len(seqn):
        # Read each frame from the webcam
        _, frame = cap.read()
        x, y, c = frame.shape

        # Flip the frame vertically
        frame = cv2.flip(frame, 1)
        # Show the final output
        cv2.imshow("Output", frame)
        if cv2.waitKey(1) == ord('q'):
            break
        frame.flags.writeable = True
        framergb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        result = hands.process(framergb)

        landmarks = []
        if result.multi_hand_landmarks:

            for handslms in result.multi_hand_landmarks:
                for lm in handslms.landmark:

                    '''lmx = int(lm.x * x)
                    lmy = int(lm.y * y)
                    lmz = int(lm.z * z)'''
                    # print(str(lmx)+" "+str(lmy))
                    landmarks.append([lm.x, lm.y, lm.z])
                    lm = calcLandMDist(landmarkGen2(frame))
                    
                    if len(seqn) == curr_seq:
                        auth = True
                        print('Authenticated')
                        break
                    else:
                        print(seqn[curr_seq])
                    if lm == seqn[curr_seq]:
                        curr += 1
                    if curr > beat:
                        curr_seq += 1
                        curr = curr % beat
                    mpDraw.draw_landmarks(
                        frame, handslms, mpHands.HAND_CONNECTIONS)
                    cv2.imwrite('output3.jpg', frame)
main()

cap.release()
cv2.destroyAllWindows()
```
